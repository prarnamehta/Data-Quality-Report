## Dataset 1: Credit

```python
# Explore the dataset we are going to use for the analysis
credit_df = pd.read_csv('Credit.csv')
credit_df
```

```python
# classify columns into categorical and quantitative variables

def check_variables(credit_df: pd.DataFrame):
    cols = credit_df.columns
    quantitative_cols = credit_df._get_numeric_data().columns
    categorical_cols = list(set(cols) - set(quantitative_cols))
    return categorical_cols, list(set(quantitative_cols))

cat_cols, quan_cols = check_variables(credit_df)
print(f'categorical columns: {cat_cols}')
print(f'quantitative columns: {quan_cols}')
```
